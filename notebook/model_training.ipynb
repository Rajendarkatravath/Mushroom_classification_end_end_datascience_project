```python
# target varaiable
y=df.iloc[:,1]
```

```python
#feature engineering
from sklearn.preprocessing import LabelEncoder

#Assuming y is the target variable
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X=pd.get_dummies(X)
X
```

```python
# spliting the dataset for training and testing
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=40)
```

```python
X_train
```

```python
# all models in one code
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
import warnings

# Define the evaluation function
def evaluate_model(true, predicted):
    cm = confusion_matrix(true, predicted)
    classification_reports = classification_report(true, predicted)
    acc_score = accuracy_score(true, predicted)  # Rename the variable here

    return cm, classification_reports, acc_score



# Define the models
models = {
    "Logistic Regression": LogisticRegression(C=0.001, penalty='l2', solver='liblinear'),
    "SVM Classifier": svm.SVC(kernel='linear', C=0.0001),
    "Decision Tree": DecisionTreeClassifier(criterion='entropy', max_depth=1, min_samples_leaf=5),
    "Random Forest Classifier": RandomForestClassifier(n_estimators=50, max_depth=1, min_samples_leaf=5),
    "Gradient Boosting Classifier": GradientBoostingClassifier(n_estimators=15, max_depth=1, min_samples_leaf=3, learning_rate=0.1),
    "AdaBoost Classifier": AdaBoostClassifier(n_estimators=2, learning_rate=1.0, algorithm='SAMME')

}

# Iterate over the models
for name, model in models.items():
    model.fit(X_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluate Train and Test dataset
    cm_train, cr_train, as_train = evaluate_model(y_train, y_train_pred)
    cm_test, cr_test, as_test = evaluate_model(y_test, y_test_pred)

    print(name)

    print('Model performance for Training set')
    print("- Confusion Matrix:\n", cm_train)
    print("- Classification Report:\n", cr_train)
    print("- Accuracy Score: {:.4f}".format(as_train))

    print('----------------------------------')

    print('Model performance for Test set')
    print("- Confusion Matrix:\n", cm_test)
    print("- Classification Report:\n", cr_test)
    print("- Accuracy Score: {:.4f}".format(as_test))
    print('\n')
```
